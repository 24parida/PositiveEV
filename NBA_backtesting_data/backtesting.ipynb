```python
import requests
import urllib
import datetime
import json


BASE_URL = 'https://api.prop-odds.com'

class myException(Exception):
    pass

def get_request(url):
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    
    if response.status_code == 401:
        raise myException("ERROR 401")
    raise myException(f"ERROR {response.status_code}")

def get_nba_games(date, api):
    query_params = {
        'date': date,
        'tz': 'America/New_York',
        'api_key': api,
    }
    params = urllib.parse.urlencode(query_params)
    url = BASE_URL + '/beta/games/nba?' + params
    return get_request(url)


def get_most_recent_odds(game_id, market, api):
    query_params = {
        'api_key': api,
    }
    params = urllib.parse.urlencode(query_params)
    url = BASE_URL + '/beta/odds/' + game_id + '/' + market + '?' + params
    return get_request(url)


def get_fantasy_lines(game_id, market, api):
    query_params = {
        'api_key': api,
    }

    params = urllib.parse.urlencode(query_params)
    url = BASE_URL + '/beta/fantasy_lines/' + game_id + '/' + market + '?' + params
    return get_request(url)

```
